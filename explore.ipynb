```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_swiss_roll
```

# Forward diffusion process

A Gaussian kernel is applied at every step. Given that T is sufficiently large and $\beta$ is considerably small, the final distribution will be a well behaved isotropic gaussian.

$$q(x^{(t)}|x^{(t+1)} = \mathcal{N}\left(x^{(t)}; \sqrt{1 - \beta_t}x^{(t-1)}, \mathbb{I}\beta_t\right)$$

```python
T = 150

points = make_swiss_roll(n_samples=300)[0][:, [0,2]]

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(3*3, 3))
ax1.set_axis_off(), ax2.set_axis_off(), ax3.set_axis_off()
ax1.plot(points[:, 0], points[:, 1], color='C0', marker='o', ls='', alpha=0.7)

for t in range(T):
    beta = 0.05
    points = np.random.normal(loc=np.sqrt(1-beta)*points, scale=beta, size=points.shape)
    if t == 50:
        ax2.plot(points[:, 0], points[:, 1], color='C0', marker='o', ls='', alpha=0.7)

ax3.plot(points[:, 0], points[:, 1], color='C0', marker='o', ls='', alpha=0.7)
plt.show()
```
